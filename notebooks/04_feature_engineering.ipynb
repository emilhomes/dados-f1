```python
import pandas as pd

df = pd.read_csv('../data/raw/races_2023.csv')

#aplicando one-hot-enconding nos pilotos e equipes
df_encoded = pd.get_dummies(df, columns=['Driver', 'Constructor'], prefix=['Driver', 'Team']) 

columns_to_drop = [
    'Name', 'Date', 'Url',
    'Status', 'Points', 'Time_Millis',
    'Fastest_Lap_Rank', 'Fastest_Lap_Time'
]

df_final = df_encoded.drop(columns_to_drop, axis=1)

df_final.to_csv('../data/processed/f1_processed.csv', index=False)
```
